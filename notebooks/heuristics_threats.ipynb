```python
import sys, os
sys.path.append(os.path.dirname(os.path.abspath('')))
from src.hueristic import Hueristic
from src.utils import display
import chess
from chess import SquareSet as ss

sep = '-'*20
f = Hueristic()
b = chess.Board(fen = 'B3r3/3p4/2b3p1/7Q/5N2/8/8/k3R3 w - - 0 1')
c = chess.WHITE
display(b)
f.set(b)


# hanging pieces
for p in chess.PIECE_TYPES:
    print(p, chess.PIECE_NAMES[p])
    for p_ in f.pieces[c][p]:
        sqset = ss([p_])
        print(sqset, '\n')
        print('hanging : ', f._hanging(sqset, p_, not c))
    
    for p_ in f.pieces[not c][p]:
        sqset = ss([p_])
        print(sqset, '\n')
        print('hanging: ', f._hanging(sqset, p_, c))
```
