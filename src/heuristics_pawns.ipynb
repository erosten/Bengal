```python
from ai.evaluations.hueristic import Hueristic
from utils import display
import chess
from chess import SquareSet as ss

sep = '-'*20
f = Hueristic()
b = chess.Board(fen = '8/8/4pPP1/1p2P3/1P6/1P6/8/8 w - - 0 1')
c = chess.WHITE
display(b)
f.set(b)
```

```python

for pawn in f.pieces[c][chess.PAWN]:
    print(ss([pawn]), '\n')
    print('Blocked: ', f._blocked(ss([pawn]), c))
    print('Doubled: ', f._doubled(ss([pawn]), c))
    print('Doubled isolated: ', f._doubled_isolated(pawn, ss([pawn]), c))
    print('Isolated: ', f._isolated(pawn, c))
    print('phalanx: ', f._phalanx(ss([pawn]), c))
    print('supported: ', f._supported(ss([pawn]), c))
    print('opposed: ', f._opposed(ss([pawn]), c))
```
