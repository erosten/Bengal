```python
from ai.evaluations.hueristic import Hueristic
from utils import display
import chess
from chess import SquareSet as ss

sep = '-'*20
f = Hueristic()
b = chess.Board(fen = 'b6n/p6p/3PPkb1/2nBN3/2p2p2/5K2/P5BP/N4R2 w - - 0 1')
c = chess.WHITE
display(b)
f.set(b)

print('White')
print('Minor pieces behind pawn (3)', f._minors_behind_pawn(c))
print('Rook on Q (F): ', f._rook_on_queen_file(c))
print('Rook on Kring (T): ', f._rook_on_king_ring(c))
print('Bishop on KR (T): ', f._bishop_on_king_ring(c))
print('\n\n Black')
print('Minor pieces behind pawn (3)', f._minors_behind_pawn(not c))
print('Rook on Q (F): ', f._rook_on_queen_file(not c))
print('Rook on Kring (T): ', f._rook_on_king_ring(not c))
print('Bishop on KR (T): ', f._bishop_on_king_ring(not c))
```

```python
# rooks
f = Hueristic()
b = chess.Board(fen = '8/1P6/8/2p1rr2/r3p3/5P2/8/1RR4R w - - 0 1')
c = chess.BLACK
display(b)
f.set(b)

for c in [True, False]:
    print('white' if c else 'black')
    for rook in f.pieces[c][chess.ROOK]:
        print(ss([rook]), '\n')
        print('Semi/open or open file (0,1,2): ', f._rook_on_open_or_semi_open_file(rook, c))
```

```python
# x - ray masks
f = Hueristic()
b = chess.Board(fen = 'rnbqkbnr/pppppppp/8/8/8/8/PPPPPPPP/RNBQKBNR w KQkq - 0 1')
c = chess.BLACK
display(b)
f.set(b)

for c in [True,False]:
    print('bishops')
    for bishop in f.pieces[c][chess.BISHOP]:
        print(ss([bishop]), '\n')
        print(f._get_bishop_xray_mask(bishop))

    print('rooks')
    for rook in f.pieces[c][chess.ROOK]:
        print(ss([rook]), '\n')
        print(f._get_rook_xray_mask(rook))
```
